```python
import numpy as np
```

```python
arr=np.array([1,3,2,3,4,2,1,5])
```

```python
print("Max:",np.max(arr))
```

```python
print("Min:",np.min(arr))
```

```python
print("Argmax:",np.argmax(arr))
```

```python
print("Argmin:",np.argmin(arr))
```

```python
print("Unique:",np.unique(arr))
```

```python
print("Bincount:",np.bincount(arr))
```
